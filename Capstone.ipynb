# **Function**

```python
nltk.download('words')
nltk.download('punkt')
def remove_non_english_words(text):
    english_vocab = set(words.words())
    words_list = nltk.word_tokenize(text)
    english_words = [word for word in words_list if word.lower() in english_vocab]
    modified_text = ' '.join(english_words)
    return modified_text
```

```python
def separate_compound_words(text):
    separated_words = []
    words = text.split()
    for word in words:
        separated_words.extend(wordninja.split(word))
    separated_sentence = ' '.join(separated_words)
    return separated_sentence
```

```python
dataset1 = pd.read_csv(ds1)

nltk.download('stopwords')
stop_words = stopwords.words('english')
dataset1['text_nostop'] = dataset1['full_text'].apply(lambda x: ' '.join([word for word in str(x).split() if word not in (stop_words)]))
dataset1
```

```python
lemmatizer = WordNetLemmatizer()
dataset1['text_lemmatized'] = dataset1['text_nostop'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in nltk.word_tokenize(x)]))
dataset1
```

```python
dataset1['text_tokenized'] = dataset1['text_lemmatized'].apply(lambda x: str.split(x))
dataset1
```
